# Visulaize the borough that contains toronto in the name

```python
<h4> Extracting
```

```python
old_list=list(df['Borough'].unique())
```

```python

r=re.compile(".*Toronto")
newlist=list(filter(r.match,old_list))
newlist


```

```python

a=(df['Borough'] =='Downtown Toronto') | (df['Borough'] =='East Toronto') | (df['Borough'] =='West Toronto') | (df['Borough'] =='Central Toronto')
sub_df=df[a]

```

```python

address = 'Canada,CA'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
lat = location.latitude
lon = location.longitude
```

```python
map_canada = folium.Map(location=[lat, lon], zoom_start=11)

for i,j,neigh in zip(sub_df['Latitude'],sub_df['Longitude'],sub_df['Neighborhood']):
    neigh=folium.Popup(neigh,parse_html=True)
    folium.CircleMarker(
        [i,j],
        radius=5,
        popup=neigh,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_canada)  
    
map_canada
    
```
